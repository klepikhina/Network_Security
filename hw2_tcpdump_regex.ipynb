```python
#Need to learn a better way than do a system call
#Email
filename = '/home/klepikhina/Documents/Homework/Homework2/tcpdump_hw2.txt'
if not os.path.isfile(filename):
    os.system("strings tcpdump_hw2 > tcpdump_hw2.txt")
    filename = '/home/klepikhina/Documents/Homework/Homework2/tcpdump_hw2.txt'

emailList = []
page404List = []
with open(filename) as f:
    data = f.readlines()
    for line in data:
        line = line[:-1]
        email = re.search("([a-zA-Z0-9_.+-]+@[a-zA-z0-9-]+\.[a-z]+$)", line)
        page404 = re.search(" 404 ", line)
        if email is not None:
            email = str(email)
            splitAt = "match='"
            saveEmail = email[(email.index(splitAt)+(len(splitAt)+1)):-2]
            emailList.append(line)
        if page404 is not None:
            page404List.append(line)
    emailString = ', '.join(emailList)
   

#IMG tag
packetcapture = '/home/klepikhina/Documents/Homework/Homework2/packetcapture.txt'
if not os.path.isfile(packetcapture):
    os.system("tshark -r tcpdump_hw2 -V > packetcapture.txt")
    packetcapture = '/home/klepikhina/Documents/Homework/Homework2/packetcapture.txt'
IMGList = []
with open(packetcapture) as f:
    data = f.readlines()
    for line in data:
        line = line[:-1]
        img = re.search("<img ", line)
        if img is not None:
            img = str(img)
            IMGList.append(img)

#For CNAME and A records
test = '/home/klepikhina/Documents/Homework/Homework2/tcpdump_hw2_test.txt'
if  not os.path.isfile(test):
    os.system("tshark -r tcpdump_hw2 > tcpdump_hw2_test.txt")
    test = '/home/klepikhina/Documents/Homework/Homework2/tcpdump_hw2_test.txt'
cList = []
aList = []
with open(test) as f:
    data = f.readlines()
    for line in data:
        line = line[:-1]
        c = re.search("CNAME\s*((([^.]+.+[^.])\{1,10\})|\S+)", line)
        a = re.search("\sA\s(((\d|\d{2})|\d{3})+.+((\d|\d{2})|\d{3})+.+((\d|\d{2})|\d{3})+.+((\d|\d{2})|\d{3}))\S", line)
        if c:
            c = str(c)
            cList.append(c[48:-2])
        if a:
            a = str(a)
            aList.append(a[48:-2])
    cString = ' \n'.join(cList)
    aString = '\n'.join(aList)

print("The emails found are: \n" + emailString)
print("C records in PCAP: \n" + cString)
print("A records in PCAP: \n" + aString)
print("There are " + str(len(IMGList)) + " IMG tags.")
print("There are " + str(len(page404List)) + " 404 pages.")


#Email that is found is from the people that are doing the encryption...
```
